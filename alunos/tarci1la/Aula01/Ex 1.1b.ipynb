```julia
using JuMP
using HiGHS

# Matriz de pontuações
S = [
    90 80 10 50
    60 70 50 65
    70 40 80 85
]

# Horas requeridas por projeto
r = [70, 50, 85, 35]2

n_designers = 3
n_projects = 4
horas_max = 80

model = Model(HiGHS.Optimizer)

# Variáveis de decisão
@variable(model, x[1:n_designers, 1:n_projects] >= 0)

# Função objetivo
@objective(model, Max, sum(S[i,j] * x[i,j] for i in 1:n_designers, j in 1:n_projects))

# Restrição de disponibilidade dos designers
@constraint(model, [i in 1:n_designers], sum(x[i,j] for j in 1:n_projects) <= horas_max)

# Restrição de horas exigidas por projeto
@constraint(model, [j in 1:n_projects], sum(x[i,j] for i in 1:n_designers) == r[j])

# Resolver
optimize!(model)

println("Status da solução: ", termination_status(model))
println("Valor ótimo da função objetivo = ", objective_value(model))
println()

println("Matriz de alocação ótima:")
for i in 1:n_designers
    for j in 1:n_projects
        println("x[$i,$j] = ", value(x[i,j]))
    end
end

println()
println("Resumo por designer:")
for i in 1:n_designers
    println("Designer $i: ", sum(value(x[i,j]) for j in 1:n_projects), " horas")
end

println()
println("Resumo por projeto:")
for j in 1:n_projects
    println("Projeto $j: ", sum(value(x[i,j]) for i in 1:n_designers), " horas")
end
```
